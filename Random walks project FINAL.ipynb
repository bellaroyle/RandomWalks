We start off by creating a random walk function to ensure we do not have to hard code this every time we don't want a random walk.<br>

Instead we call all the the $\theta$ values at once and use NumPy's cumulative sum feature to calculate this random walk in 2 steps. Our function takes in variables T, show and delta. T is the amount of time steps in the walk, show is a boolean variable that allows you to plot the walk and delta is the spatial unit.
The output is a (T+1) x 2 matrix containing the random walk of T steps.

```python
def RandomWalk(T, show = True,  delta = 1,):
    """
    Vectorised random walk function.
    """
    r = np.zeros((T + 1,2))
    theta = np.random.uniform(0, 2*np.pi, size = T)  
    
    # Starts from (0,0), so start from r[1]
    r[1:]=np.array([np.cos(theta)*delta,np.sin(theta)*delta]).T
    r=np.cumsum(r,axis=0)                                         
    
    # Shows the plot of the random walk unless false is inputted into the function
    if show == True:                                              
        plt.figure(figsize = (18,12))
        plt.plot(r[:,0], r[:,1], '-b')
        plt.plot(0, 0, 'go', label = 'Start')
        plt.plot(r[-1,0], r[-1,1], 'ro', label = 'End')
        plt.xlabel('X',fontsize=16)
        plt.ylabel('Y',fontsize=16)
        plt.title('2D Random Walk of {:,} Steps'.format(T), fontsize=14, fontweight='bold')
        plt.grid(True)
        plt.legend()
        plt.show()
    
    
    return r                                                       
```

## Average Walk Function

This code gives a clear idea of what is happening in the random walks on average. The function loops N times and populates 2 separate matrices for N sets of $x$ and $y$ values, forming 2 $ T+1 × N$ matrices. It then averages across each row for the average position at each time step in the $x$ and $y$ axis. The $x$ and $y$ arrays are then joined together to form the average walk. This is also done for the standard deviation of the average walk. NumPy automatically uses a new seed each iteration so we do not need to seed the function. 

We have added the boolean variable 'export' to allow you to choose whether you export each one of the random walks to a CSV file. They are named 'Random_walk_iteration' followed by the number of the iteration. 

The output of the function is 2 matrices containing averages and standard deviations at each time step in the $x$ and $y$ axis.

```python
def AverageWalk(N,T,export=False):
    '''
    This creates an average over N
    simulations of a random walk with T steps.
    '''
    x = np.empty((T + 1, N))
    y = np.empty((T + 1, N))
    Average = np.empty((T + 1, 2))
    
    # Loop through creating N columns of x and y values
    # Each time we call the original function to ensure a new seed
    for i in range(0,N):
        Walk = RandomWalk(T,False)
        x[:,i] = Walk[:,0]
        y[:,i] = Walk[:,1]
    # If you input True, the code saves each run in a separate csv file 
        if export:
            title = 'CSV\Random_walk_iteration' + str(i + 1) + '.csv'
            np.savetxt(title,Walk,delimiter = ',')
    
    # Calculate the mean and standard deviation across the rows for each time step
    averagex = np.mean(x,axis = 1) 
    averagey = np.mean(y,axis = 1)
    
    StdX = np.std(x,axis = 1) 
    StdY = np.std(y,axis = 1)
    
    # Combine the columns and export it as a T x 2 matrix
    Average = np.column_stack((averagex,averagey))
    Std = np.column_stack((StdX,StdY))
    return Average,Std
```

Compute the mean square displacement (MSD) of your trajectory. This means the following: Formally, the MSD is the statistical average of the square distance the particle moves in a time $\delta t$, i.e.

$$MSD(\delta t) = ⟨|r(t + \delta t) − r(t)|^2⟩_{t,ensemble}$$

Here the ensemble average means that the MSD is the mean over different runs of your system (you have that data saved already). The t-average means that you will need to average over all t that exist such that both $0≤t<T$ and $0≤t+δt<T$. In detail,this means:

$$⟨|r(t + δt) − r(t)|^2⟩_t = \frac{1}{N_t} \sum_{k} |r(k + δt) − r(k)|^2$$

where $N_t$ is the number of k’s such that this condition is fulfilled.

You can have your code compute $N_t$ directly as you do your sum. It is possible to use NumPy slicing together with np.mean(...) to do the sum in a single step; this is also faster than a loop

### Logic behind our code:
We use NumPy's slicing function to calculate the mean square displacement (MSD).

__First case:__<br>
If $\delta t =0 $, the MSD is zero because a particle in a random walk will not have moved in that time.

__Second case:__<br>
If $\delta t<0$ or $\delta t\ge T$, an error message is displayed as $\delta t$ is out of range.

__Third Case:__<br>
When $0<\delta t<T$, the MSD can be calculated using the formulas given. 



The function 'MSD()' takes a random walk and $\delta t$ as inputs.
The random walk is split into two column vectors containing  the $x$ and $y$ coordinates of the walk.

We subtract an array containing points 0 to T- $\delta t$ from an array containing points T to $\delta t$  then square this for the MSD for both $x$ and $y$ coordinates using NumPy's slicing features, then the difference of these two arrays is the change in position that has occurred in the time $\delta t$ (displacement). We then add both arrays together and take the mean which gives us the MSD. The function returns this value.


```python
def MSD(rwalk,dt):
    '''
    This function calculates the mean squared displacement (MSD) using 
    numpy slicing.
    '''
    T = len(rwalk) - 1
    # Dt must be between 0 and T
    if dt == 0:
        return 0
    if 0 < dt < T:
        rx = rwalk[:,0]        
        ry = rwalk[:,1]
        # Array slicing to do the subtraction
        # Square each element in the matrices
        displacementx = np.square(rx[dt:]-rx[:-dt])
        displacementy = np.square(ry[dt:]-ry[:-dt])
        
        # Add both arrays together
        displacement = np.add(displacementx,displacementy)
        y = np.mean(displacement)
        
        return y
        
    else:
        return "dt value not in correct range"
  
```

### Logic behind our code:


Our function takes inputs of $N$ and $T$ where $N$ is the number of runs and $T$ is the steps of a random walk.


The first for loop calculates a random walk of T steps using our 'RandomWalk()' function. The second for loop calculates the MSD using our 'MSD()' function for each value of $\delta t$ at each step of the walk storing it in Msd_Vals. This is repeated for N different random walks generated. 
Msd_Vals is $ T × N$ array with the MSD for each $\delta t$ stored in columns, each new column being a different random walk.

The standard deviation of the MSD is calculated along the rows of the array Msd_Vals, which gives the standard deviation of MSD at each $\delta t$. We use the formula given to calculate the error bars.
We then apply np.mean along the rows of the array Msd_Vals to obtain the mean value of MSD at each $\delta t$.

Finally, we plot these values of the MSD against the values of $\delta t$ with error bars and calculate explicitly the gradient of the graph so it can be printed.

```python
def MSD_Plot(N,T):
    '''
    Here we are plotting the MSD as a function of δt.
    '''
    δt  = np.arange(0,T)
    Msd_Vals = np.zeros((T,N))
    msd_std = np.zeros((T,))
    
    # Loop which calculates MSD at each δt for N random walks
    for u in  range(N):
        rwalk = RandomWalk(T,show=False)
        for i in range(0,T):
            Msd_Vals[i,u] = MSD(rwalk,δt[i])
            
    # Calculates standard deviation of each row of MSD_Vals
    msd_std = (np.std(Msd_Vals,axis = 1))
    err_bar = msd_std/np.sqrt(N)
    # Calculates mean MSD at each dt by calculating mean of each row of MSD_Vals
    ensemble_mean = np.mean(Msd_Vals,axis = 1)
    
    # Plot MSD as a function of δt 
    plt.figure(figsize = (8,8))
    plt.xlabel('$\delta t$',fontsize=16)
    plt.ylabel('MSD($\delta t$)',fontsize=16)
    plt.title('Mean squared displacement as a function of δt', fontsize=14, fontweight='bold')
    f = plt.errorbar(δt , ensemble_mean, xerr=None, yerr=err_bar , ecolor = 'cyan', fmt='-.', label = "MSD with error bars")
    g,v = np.polyfit(δt ,ensemble_mean,deg = 1) # Two outputs given but we want g
    print('The gradient of the graph is ' + str(g))
    plt.plot(δt ,δt ,'-r', label = "$\Delta$")
    plt.legend()
    plt.grid()
    plt.show()
    return f
```

## Persistent Random Walk 

In a persistent random walk, the particle does not completely change direction at every time step. Simulate a persistent random walk using the equations:
$$r(i+1) = r(i) + (cos (\theta(i)), sin (\theta(i)))\Delta$$
$$\theta(i+1) = \theta(i) + D_r \eta(i)$$
where $\eta(i)$ is chosen from a standard normal distribution $\mathcal{N}(0,1)$. 

### Logic behind our code:
Our function takes four inputs:
<br/>__Dr__: Controls the range that $θ$ can be in at each step. <br/>__T__ : 
The number of steps of the persistent walk. <br/>__show__ : 
Boolean variable controlling output of the graph.<br/>__delta__: Spatial unit.

Theta is an array representing the $D_r \eta(i)$ values in the formula. We then apply np.cumsum to the array theta which outputs the cumulative sum of the values in the array, this is now an array containing all the $\theta(i)$ values.

The values of $(cos (\theta(i)), sin (\theta(i)))\Delta$ are calculated as arrays and transposed, so that these values can now be stored in our array r. We then apply np.cumsum to the array r which outputs the cumulative sums of the values, this is now an array containing all our values of $r(i)$.

Finally we plot the $x$ and $y$ coordinates of the persistent random walk against each other.

```python
def pers_rw(Dr, T, show = True,  delta = 1):
    '''
    This function takes the input Dr and T to create a random walk
    that does not completely change direction at each step
    '''
    
    r = np.zeros((T + 1,2))
    n = np.random.normal(0,1, size = T)
    theta = Dr*n
    theta = np.cumsum(theta,axis = 0)
    
    r[1:] = np.array([np.cos(theta)*delta,np.sin(theta)*delta]).T
    r = np.cumsum(r,axis=0)  
        
    # Plots the trajectory unless False is inputted to function
    if show == True:
        plt.figure(figsize = (10,7))
        plt.plot(r[:,0], r[:,1], '-m')
        plt.xlabel('X',fontsize=16)
        plt.ylabel('Y',fontsize=16)
        plt.title(r'2D Persistent Random Walk of {:,} Steps with $D_r$={}'.format(T,Dr), fontsize=14, fontweight='bold')
        plt.grid(True)
        plt.plot(0, 0, 'go', label = 'Start')
        plt.plot(r[-1,0], r[-1,1], 'ro', label = 'End')
        plt.legend()
        plt.show()

    return r
```

For $D_r ≤ 1$, the full curve should follow:

$$ MSD(\delta t) =\frac{4\Delta}{D_r^2}\biggl( \delta t + \frac{2\Delta}{D_r^2}(e^{\frac{−D_r^2t}{2\Delta}} - 1 )\biggr)     \qquad    (*) $$

```python
def pers_MSD(Dr, T, N = 10, delta = 1):
    '''
    This function plots the MSD of fixed N = 10 persistent random walks 
    each of T steps with input Dr.
    '''
    P = delta/(Dr)**2
    y = np.zeros((T,N))
    x = np.arange(0,T)
    
    # For loop which calculate MSD at each δt for each N = 10 persistent walks
    for j in range(N):
        rw = pers_rw(Dr,T,False, delta = 1) 
        for i in range(T):
            y[i,j] = MSD(rw,x[i])
            
    # Calculates the mean of MSD at each δt by calculating mean of each row of y
    ensemble_mean = np.mean(y,axis = 1)
    
    # Calculates the values of (*) for each δt value in x 
    msd=4*P*(x + 2*P*(np.exp((-x*Dr**2)/2*delta)-1))
 
    
    
    plt.figure(figsize = (10,7))
    plt.plot(x,ensemble_mean,'-o', label = "MSD")
    plt.plot(x,msd,'-r', label = "(*)")
    plt.axvline(x = 1/Dr**2, linestyle = 'dotted', color = 'k',label = '$1/Dr^2$')
    plt.title('MSD of persistent random walk of {:,} steps with $D_r$={} and $\Delta$={}'.format(T,Dr,delta),
              fontsize = 14, fontweight = 'bold')
    plt.xlabel('δt',fontsize=16)
    plt.ylabel('MSD(δt)',fontsize=16)
    plt.legend()
    plt.grid(True)
    plt.show()
  
    return ensemble_mean, msd
```

## First Passage time
Suppose that you, a random particle staggering out of a bar next to a sheer cliff face at
1am on a Sunday morning, are attempting to make it back to your flat but you are not
successful. Starting from the bar located at $(0,0)$ you instead hit the cliff, a north-south
line located at $x = L$, and you very unfortunately fall down. Calculate your mean life
expectancy $\tau(L)$ before you to meet your end, as a function of the distance L of the beach.
Reasonable starting values would be $dt = 1s$ as before, $\Delta = 1m$, and $L = 5m$. This type
of time to reach a certain distance or threshold is called a first passage time.


Since we are only looking at x, this problem is a 1D random walk starting from an initial position $r(0,0) = (0,0)$ with equation $ r(i+1) = r(i) + (\cos \theta_i)∆$.

In this case we will not plot the random walk so we omit the first point (0,0).

We initialise 3 count values r, i and j and create a random walk of length T.
We then use a while loop setting r to the last value of the random walk, if $r>L$ the loop breaks and returns i,the count of total amount of steps taken.

To avoid repetitively calling np.random.uniform(0, 2$\pi$) we call T values at the same time, when we reach the end of this array $j=T$ then we call another T values and concatenate them to the end of the array



Since the first passage time is a random function, $N=T(L)$ could be arbitratily large and tend to infinity. Since this is the case we need to ensure our function is as fast as possible in calculating N. Becasuse of this we have optimised the function using Cython which is " designed to give C-like performance with code that is written mostly in Python with optional additional C-inspired syntax."

```python
%load_ext Cython
```

```python
%%cython


import numpy as np
cimport numpy as cnp
cimport cython

@cython.boundscheck(False)  # Deactivate bounds checking
@cython.wraparound(False)   # Deactivate negative indexing.
def FptCython(L): 
    """
    A function to calculate the first passage time to point x=L
    It returns single value i, the amount of time passed before hitting point L
    
    """
    # Similar to C we must define variable types when they are initialised
    cdef double r=0
    cdef int i=0
    cdef int j=0
    cdef int T=100000
    
    # Creating empty arrays for theta and extravalues
    cdef cnp.ndarray[cnp.float64_t , ndim=1] theta,extravalues
    theta = np.cumsum(np.cos(np.random.uniform(0, 2*np.pi, size = T)))
    while r<=L:
        if j == T:
            j=0
            # This if statement adds more theta values after we have looped through all the current values
            extravalues=np.cumsum(np.cos(np.random.uniform(0, 2*np.pi, size = T)))
            extravalues=extravalues+theta[i-1]
            theta=np.concatenate((theta, extravalues), axis=0)
        # Update the r values and increase the counts by 1 top keep the while loop going
        r=theta[i]
        i+=1
        j+=1
    return i
```

```python
%%cython


import numpy as np
cimport numpy as cnp
cimport cython

@cython.boundscheck(False)  # Deactivate bounds checking
@cython.wraparound(False)   # Deactivate negative indexing.
def firstpassagetimecython_nmax(L,nmax): 
    '''
    This function is identical to firstpassagetimecython function, but takes in value nmax, which is the maximum value 
    'i' can take
    '''
    # Similar to C we must define variable types when they are initialised
    cdef double r=0
    cdef int i=0
    cdef int j=0
    cdef int T=100000
    
    # Creating empty arrays for theta and extravalues
    cdef cnp.ndarray[cnp.float64_t , ndim=1] theta,extravalues
    
    theta = np.cumsum(np.cos(np.random.uniform(0, 2*np.pi, size = T)))
    while r<=L and i<nmax:
        if j == T:
            j=0
            # This if statement adds more theta values after we have looped through all the current values
            extravalues=np.cumsum(np.cos(np.random.uniform(0, 2*np.pi, size = T)))
            extravalues=extravalues+theta[i-1]
            theta=np.concatenate((theta, extravalues), axis=0)
        # Update the r values and increase the counts by 1 top keep the while loop going
        r=theta[i]
        i+=1
        j+=1
    return i
```

```python
def meanfirstpassagetime(L,N,nmax):
    '''
    Function to calculate mean first passage time after N runs with parameters (L,nmax)
    '''
    r=[firstpassagetimecython_nmax(L,nmax) for x in range(N)]
    return np.mean(r)


def plot_meanpasstime(Start,Stop,N,L):
    '''
    Function to plot the mean first passage time as a function of nmax values
    where the nmax values range from the inputs 'Start' to 'Stop'
    '''
    # x is values from Start to Stop with step = 1000
    x=range(Start,Stop,1000)
    length=len(x)
    y=np.zeros((length,2))
    z=0
    # For loop to assign row values of y as nmax and mean first passage time with input nmax
    for i in x: 
        y[z,:] = i,meanfirstpassagetime(L,N,i) 
        z+=1
    plt.figure(figsize = (10,7))
    plt.xlabel('Nmax',fontsize=16)
    plt.ylabel(r' $\tau(L)$',fontsize=16)
    plt.title('Mean life expectancy for L={} and {} Runs'.format(L,N))
    b=plt.plot(y[:,0], y[:,1],'-ro')
    plt.grid()
    plt.show()
    return 
```

We can look at mean first passage time's in different shapes with different rules. These shapes all have $\mathbb{E}{(T(X))}<\infty$ so they can be plotted.

1. Circular cliff with the bar centred at $(0,0)$
2. Rectangular cliff with the bar centred at $(0,0)$
3. Infinite strip of land (with the choice of x or y axis) with the bar at $(0,0)$

We use the same logic as we did in previous functions with a minor changes. Again we use count $i$ and $j$, but since we start at $(x_0,y_0)$ as opposed to $(x_1,y_1)$ , $j$ starts at -1, and we subtract 1 from $i$ at the end of the code.

__Circle__

We take input value L. Using the distance formula we can see  $(x,y)$ is inside a circle of radius L if $ x^2+y^2 \lt L^2$, $(L>0)$. 

__Rectangle__

### Random Walk inside box with an open lid

Finally we create a random walk that is restricted to a box with an open lid. The difference being that the point can hit the walls and continue the random walk inside the box.


 $x_1<x<x_2$ and $y_1<y$

In this case for points $(x,y)$ the conditions for the while loop to continue this walk are $y<y_2$.


We assume the point travels a distance of 1 unit every step of the random walk, so if it was to collide with the wall after travelling distance $a$, it would only travel $1-a$ before another random walk step is added.

We use the same logic as previously described the only change being that when any combination of the conditions  
$x_1>x , x_2<x, y_1>y$ whilst $y<y_2$ we need to reflect the points back into the shape, add new points, delete old ones, then change the new points (since our random walk is pre populated with T $(x,y)$ points). 

We have created a function 'boundarybounce()' that deals with all cases like this. Contained within this function are 4 other functions, 'boundarybelow()', 'boundaryright()' and 'boundaryleft()' all reflect points at walls at the bottom, right and left respectively.

Distance measures the distance between two walls $y=y1$ and $x=x1$ or $x=x2$ so it can decide what wall the particle will hit first. We have to take the particles trajectory into account when doing this so first we calculate gradients to check what wall it will hit first. If it will hit the wall at the bottom we call boundary bottom first then boundary left second, likewise for all other conditions.

If we have values $ \left[\begin{matrix} x_{n-1} & y_{n-1} \\ x_{n} & y_{n} \end{matrix}\right] $ <br> 
where $(x_{n}, y_{n})$ is out of the range 'boundarybounce()' function will add $k$ more points till the new points are in range.



__Example__<br> 
We have a snapshot into our random walk, we are currenty at point $i$ in the random walk with $x1=y1=-1,x2=y2=1$
$ \left[\begin{matrix} x_{i-1} & y_{i-1} \\ x_{i} & y_{i} \\ x_{i+1} & y_{i+1} \end{matrix}\right] =\left[\begin{matrix} 0.5 & -0.5  \\ 1.07357644 & -1.31915204 \\ 0.08876868 & -1.14550387 \end{matrix}\right]$

Points at $i-1$ and  $i$ are passed to 'boundarybounce' function where since $x_2<x_i$ and $y_1>y_i$ the function calculates the distance to the lines $x=x_2$ and $y=y_1$ . Using this information the function then decides which wall the particle will hit first then reflects the points. Boundary bounce then returns a new array $k\ge0$<br> 
$ Boundarybounce(\left[\begin{matrix} x_{i-1} & y_{i-1} \\ x_{i} & y_{i}  \end{matrix}\right])=\left[\begin{matrix} x_{i-1} & y_{i-1} \\ a_{i} & b_{i} \\  a_{i+1} & b_{i+1}\\ \dots & \dots \\ a_{i+k} & b_{i+k} \end{matrix}\right] $<br> 
Where a,b are new x,y point output by boundarybounce. <br>

We then delete the old $(x_{i}, y_{i})$ point from the matrix, insert the new points and update count value k as we are now in position $i+k$ in the matrix, as oppsed to $i$ . In our case $k=2$ since we have added 2 extra points

```python
def distance(xprev,yprev,x,y,x1,y1):
    '''
    Calculates gradient of line from (xprev,yprev) to (x,y)
    Then outputs distances to these lines as distancex, distancey
    
    '''
    m=(yprev-y)/(xprev-x)
    c=y-m*x

    distancey=np.sum(np.subtract([(y1-c)/m,y1],[x,y])**2)
    distancex=np.sum(np.subtract([x1,m*x1+c],[x,y])**2)
    
    return distancex, distancey
```

The following functions are quite similar but just work on different walls .They calculate the gradient on the line joining 2 points together, then find the point at which this line intersects with this wall. After that it reflects the point and returns an array of the first point, the contact point with the wall and the new reflected point. There are some fail safes in the code to ensure if points are already in range they are returned as is.

Arrays in NumPy are mutable so we use the np.copy function to ensure we do not change the original array.


```python
def boundarybelow(points,y1):
    '''
    A function that reflects points where y<y1
    Outputs a new array with intermediary points

    '''
    bb_points=np.copy(points)

    # Gradient of the line
    m,c=np.polyfit(bb_points[-2:,0],bb_points[-2:,1],1)
    
    x,y=bb_points[-1,:]
    if y>=y1:
        return bb_points
    # If the previous point is the same as the boundary
    # Simply reflect the new point in the x axis
    if bb_points[-2,1]==y1:
        
        bb_points[-1]=np.array([x,y1+(y1-y)])
        return bb_points
   
    # New y coordinate for point of impact
    x0=(y1-c)/m
    # Impact point in the wall
    bb_points[-1,:]=x0,y1
    # New finishing point
    new_pointbb=x,y1+(y1-y)
    newpointsbb=np.append(bb_points,[new_pointbb],axis=0)
    return newpointsbb
    
```

```python
def boundaryright(points,x2):
    '''
    A function that reflects points where x>x2
    Outputs a new array with intermediary points

    '''
    br_points=np.copy(points)
    # Gradient of the line
    m,c=np.polyfit(br_points[-2:,0],br_points[-2:,1],1)
    x,y=br_points[-1,:]
    if x<=x2:
        return br_points

    # If the previous point is the same as the boundary
    # Simply reflect the new point in the x axis
    if br_points[-2,0]==x2:
        br_points[-1]=np.array([x2-(x-x2),y])
        return br_points

    # New y coordinate for point of impact
    y0=x2*m+c
    # Impact point in the wall
    br_points[-1,:]=x2,y0
    # New finishing point
    new_pointbr=x2-(x-x2),y
    newpointsbr=np.append(br_points,[new_pointbr],axis=0)

    return newpointsbr
    
    
```

```python
def boundaryleft(points,x1):
    '''
    A function that reflects points where x<x1
    Outputs a new array with intermediary points

    '''
    bl_points=np.copy(points)
    # Gradient of the line
    m,c=np.polyfit(bl_points[-2:,0],bl_points[-2:,1],1)
    # x point that it out of range, and y point
    x,y=bl_points[-1,:]
    
    if x>=x1 :
        return bl_points
    # If the previous point is the same as the boundary
    # Simply reflect the new point in the x axis
    if bl_points[-2,0]==x1:
        
        bl_points[-1]=np.array([x1+(x1-x),y])
        return bl_points
        
    # New y coordinate for point of impact
    y0=x1*m+c
    # Impact point in the wall
    bl_points[-1,:]=x1,y0
  
    
    # New finishing point
    new_pointbl=x1+(x1-x),y   
    newpointsbl=np.append(bl_points,[new_pointbl],axis=0)
    return newpointsbl
    
    
```

```python
def boundarybounce(points,x1,x2,y1,y2):
    '''
    A function that takes points that are out of range, and returns an array of 
    new points where values are in range

    '''
    
    bbpoints=np.copy(points)
    x,y=bbpoints[-1]
    new_points=np.copy(bbpoints)

    while (x<x1 or x>x2 or y<y1) :
        # Single case where only 1 x or y value is out of range
        if x>x2 and y>y1:
            new_points=boundaryright(bbpoints,x2)
        if x<x1 and y>y1:
            new_points=boundaryleft(bbpoints,x1)
        if y<y1 and (x1<x<x2):
            new_points=boundarybelow(bbpoints,y1)

        # Cases where both x and y are out of range    
        if y<y1 and x<x1 :
            distance1=distance(bbpoints[0,0],bbpoints[0,1],x,y,x1,y1)
            yaxisdist=distance1[1]
            xaxisdist=distance1[0]
                
            if yaxisdist>xaxisdist:
                new_points=boundarybelow(bbpoints,y1)
                new_points=boundaryleft(new_points,x1)

            else:
                new_points=boundaryleft(bbpoints,x1)
                new_points=boundarybelow(new_points,y1) 
                
        if y<y1 and  x>x2:
            distance1=distance(bbpoints[0,0],bbpoints[0,1],x,y,x2,y1)
            yaxisdist=distance1[1]
            xaxisdist=distance1[0]
                 
            if yaxisdist>xaxisdist:
                new_points=boundarybelow(bbpoints,y1)
                new_points=boundaryright(new_points,x2)
                
            else:
                new_points=boundaryright(bbpoints,x2)
                new_points=boundarybelow(new_points,y1)
                
        # Updating new points
        x,y=new_points[-1]
        bbpoints=np.copy(new_points)
        

    return new_points
        
    
```

```python
def escapesquarebounce(x1,x2,y1,y2):
    T=1000
    rwalk=RandomWalk(T,False)
    i=0
    j=-1
    x,y=rwalk[0,:]    
    k=0
    while (y<y2):   
        if j == T:
            # Add new walk values
            j=0
            rwalk=addvalues(rwalk,T)
        x,y=rwalk[i+k,:]
        
        # Conditions for hitting walls
        if x<x1 or x>x2 or y<y1:
            # Points i+k-1 to i+k
            z=i+k
            RangeVals=rwalk[z-1:z+1,:]
            NewPoints=boundarybounce(RangeVals,x1,x2,y1,y2)
            rwalk[i+k+1:,:]+=(-RangeVals[1]+NewPoints[-1])
            # Replace i+k value and insert new points
            rwalk=np.delete(rwalk,i+k,axis=0)
            rwalk=np.insert(rwalk,i+k,NewPoints[1:],axis=0)
            # Update k
            k=k+(len(NewPoints)-2)
            x,y=rwalk[i+k,:]
        i+=1
        j+=1   
        
    return i-1,rwalk[:i+k]
```

```python
def plot_meanfptime_box(Start,Stop,N):
    '''
    Function to plot the mean first passage time after N runs as a function of L
    where the L values range from the inputs Start to Stop
    '''
    
    x=range(Start,Stop+1)
    length=len(x)
    y=np.zeros((length,2))
    z=0
    # For loop which assigns row values of y
    for i in x: 
        r=[escapesquarebounce(x1=-i,x2=i,y1=-i,y2=i)[0] for z in range(N) ]
        y[z,:] = i,np.mean(r)
        # Update row value by one each loop
        z+=1
        
       
    plt.figure(figsize = (6,4))
    plt.xlabel('L',fontsize=16)
    plt.ylabel(r' $\tau(L)$',fontsize=16)
    plt.title('Mean life expectancy in an open box, {:,} runs'.format(N), fontsize=16)
    b=plt.plot(y[:,0], y[:,1],'-ro')
    plt.grid()
    plt.show()
    
    return y
```

## 3D Random Walk

Here we will plot a random walk which starts from an initial position $R(0,0,0) = (0,0,0)$ and has the local dynamics :
$$R(i+1) = R(i) + (sin θ cos φ, sin θ sin φ, cos θ)∆$$
Where $θ$ and $φ$ are random angles uniformly distributed between 0 and $\pi$ & 2$\pi$ respectively and $∆$ is a spatial unit.



```python
def ThreeDim(T,show = True, delta = 1):
    '''
    This function plots a 3D random walk of T steps
    '''
    
    R = np.zeros((T + 1 ,3))
    
    theta = np.random.uniform(0, np.pi, size = T) 
    phi = np.random.uniform(0, 2*np.pi, size = T)
    R[1:] = np.array([delta*np.sin(theta)*np.cos(phi), delta*np.sin(theta)*np.sin(phi), delta*np.cos(theta)]).T
    R = np.cumsum(R,axis=0) 
    
    if show == True:
        fig = plt.figure(figsize = (15,10))
        ax = plt.axes(projection="3d")
        X = R[:,0]
        Y = R[:,1]
        Z = R[:,2]
        ax.plot3D(X,Y,Z, '-k')
        ax.set_xlabel('X',fontsize=16)
        ax.set_ylabel('Y',fontsize=16)
        ax.set_zlabel('Z',fontsize=16)
        plt.title('3D Random Walk of ' + str(T) + ' Steps', fontsize=14, fontweight='bold')
        plt.show()
    return R
```

## Self-Avoiding Simple Random Walk
A self-avoiding random walk is a path from one point to another which never intersects itself. Self-avoiding random walks are often used by chemists in order to model the behaviour of chain-like bodies such as polymers as their physical volume prohibits multiple occupation of the same point in space. These paths are usually shown on a lattice where the path will never re-visit the same point.

For each step this code considers the directions possible for the walk to take without revisiting any of the points it has already been to. If there are possible directions the walk could take it will pick one at random and take it. If there are no possible ways it can go then it is stuck in a dead end and will terminate early. 


```python
def sarw(T,show = True):
    '''
    This function creates a Simple random walk that will not intersect itself. 
    It takes T a maximum number of steps.
    '''
    # Starts at (0,0)
    x = [0]
    y = [0]
    coordinates = set() # Set that will store all coordinates visited by the walk
    coordinates.add((0,0))
    stuck = False
    for i in range(1,T):
        directions = [(1,0), (0,1), (-1,0), (0,-1)] 
        directions_possible = [] # Stores possible directions the walk could take at each step 
        
        for (dx,dy) in directions:
            if (x[i-1] + dx, y[i-1] + dy) not in coordinates:
                directions_possible.append((dx,dy))
        
        # Checks that there is a possible direction for the walk to go in.
        # If there is, it will choose a random direction from the possible ones 
        if directions_possible: 
            dx,dy = directions_possible[np.random.randint(0,len(directions_possible))]
            coordinates.add((x[i - 1] + dx, y[i - 1] + dy))
            x.append(x[i - 1] + dx)
            y.append(y[i - 1] + dy)
            
        # If no direction is possible, walk terminates early     
        else: 
            stuck = True
            steps = i + 1
            break 
        steps = T + 1
    
    points = np.zeros((steps-1,2))
    for i in range(steps-1):
        points[i] = x[i],y[i]
        
    # Plots the walk unless False is inputted
    if show == True:
        plt.figure(figsize = (8,6))
        plt.plot(x, y, '-ob')
        plt.plot(0, 0, 'go', label = 'Start')
        plt.plot(x[-1], y[-1], 'ro', label = 'End')
        plt.axis('equal')
        plt.xlabel('X',fontsize=16)
        plt.ylabel('Y',fontsize=16)
        plt.xticks(range(min(x), max(x) + 1))
        plt.yticks(range(min(y), max(y) + 1))
        plt.grid(True)
        plt.legend()
        if stuck:
            plt.title('Self Avoiding Simple Random Walk stuck at step ' + str(steps), fontsize=14, fontweight='bold')
        else:
            plt.title('Self Avoiding Simple Random Walk of length ' + str(T), fontsize=14, fontweight='bold')
        plt.show()
    return points
```

## Animated 2D Random Walk

Another way to visualise a random walk is through an animation/movie. This way we can better visualise the random walk.
We use matplotlibs inbuilt function 'FuncAnimation()'. This essentially replots the graph over and over again adding the new points its given. 

We have added the option for the axis to be constantly updating by checking for the minimum and maximum x and y values in the array if it is closer than 0.5 to the axis, it is updated. We had to make these minimum and maximum variables global as they are defined outside our 'animate()' function


The function takes a number of inputs
    
__rwalk__: Input of random walk data

__moving_axis__: Boolean variable that allows a fixed or moving axis

__interval__: The time interval in milliseconds, between plotting new points

__repeat__: Boolean variable that allows you to repeat animation

__graphproperties__: Allows the user to input code in string format to be executed, such as new shapes to plot


```python
def animate_rw(rwalk,moving_axis=True,interval=200,repeat=True,graphproperties=''):
    '''
    A Function that takes in our random walk numpy array and animates it
    ''' 
    
    global xlim_up, xlim_dwn,ylim_up, ylim_dwn
    
    frames = len(rwalk)
    
    # Setting x and y axis limits for fixed axis as maximums and minimums of the randomwalk +-1
    max_xy = np.amax(rwalk,axis = 0) + 1
    min_xy = np.min(rwalk,axis = 0) - 1
    
    # We initially set x,y y axis limits to +-2 for the moving axis
    xlim_up = ylim_up = 2
    xlim_dwn = ylim_dwn = -2


    def animate(i):
        # Loops through our data rwalk whilst checking if the x axis and y axis are in range when plotting new points
    
        global xlim_up, xlim_dwn,ylim_up, ylim_dwn
        
        # If statement to update x and y limits on graph if we are using a moving axis
        if moving_axis == True:
            max_x,max_y = np.amax(rwalk[:i+1,:],axis = 0)
            min_x,min_y = np.min(rwalk[:i+1,:],axis = 0)
            
            if max_x >= xlim_up-.5:
                xlim_up +=1
                ax.set_xlim(xlim_dwn,xlim_up)
                
            if min_x <= xlim_dwn+.5:
                xlim_dwn -=1
                ax.set_xlim(xlim_dwn,xlim_up)
                
            if max_y >= ylim_up-.5:
                ylim_up +=1
                ax.set_ylim(ylim_dwn,ylim_up)
                
            if min_y <= ylim_dwn+.5:
                ylim_dwn  -=1
                ax.set_ylim(ylim_dwn,ylim_up)
                
        line.set_data(rwalk[:i+1,0],rwalk[:i+1,1])
        return line,
    # Used to draw a clear frame
    def init():
        line.set_data([],[])
        return line,
    
    # Initialising the plot of the graph
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    
    # Setting X and Y limits depending on whether the axis are fixed or not
    if moving_axis == True:
        ax.set_ylim(ylim_dwn,ylim_up)
        ax.set_xlim(xlim_dwn,xlim_up)
    else:
        ax.set_ylim(min_xy[1],max_xy[1])
        ax.set_xlim(min_xy[0],max_xy[0])
        
        
    # Labelling axis and plotting x,y point at 0 and a title
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.plot(0,0,'or')
    ax.grid()
    ax.set_title('Animated Simple Random Walk', fontsize=14, fontweight='bold')
    exec(graphproperties)
    line, = ax.plot([],[],lw = 2, c = 'k')
    
    # This initialises the animatiion
    anim = animation.FuncAnimation(fig, animate, init_func = init, frames = frames, 
                                   interval = interval, blit = False,repeat = repeat)
    
    # Ensure there is no display
    plt.close()
    return anim
```
